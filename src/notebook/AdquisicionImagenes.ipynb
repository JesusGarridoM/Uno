```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=[]
colores=['r','g','b','y']
for color in colores:
  for num in range(1,10):
    for i in range(1,3):
      path="/content/drive/MyDrive/CURSO_KERAS/DATA/"+color+str(num)+"_"+str(i)+".jpg"
      imgOr=cv2.imread(path)
      imgRe=cv2.resize(imgOr,(48,104))
      data.append(imgRe)
data=np.array(data)
```

```python
plt.subplot(121)
plt.imshow(cv2.cvtColor(imgOr, cv2.COLOR_BGR2RGB))

plt.subplot(122)
plt.imshow(imgRe)
```

```python
#coding:utf-8
import cv2
cap = cv2.VideoCapture(6)
flag = cap.isOpened()

index = 1
print('aun no entraa')
while(flag):
  print('Entro al while')
  ret, frame = cap.read()
  cv2.imshow("Capture_Paizhao",frame)
  k = cv2.waitKey(1) & 0xFF
  if (k =='s'):
    cv2.imwrite('m1.jpg' + str(index) + ".jpg", frame)
    print(cap.get(3))
    print(cap.get(4))
    print("save" + str(index) + ".jpg successfuly!")
    print("-------------------------")
    index += 1
  elif (k =='q'):
    break
cap.release()
cv2.destroyAllWindows()
```
